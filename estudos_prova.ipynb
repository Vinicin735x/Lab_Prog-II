```python
import numpy as np
import matplotlib.pyplot as plt

class Plotter:
    def __init__(self, data):
        self.data = np.array(data)

    def plot(self):
        plt.plot(self.data)
        plt.show()

# dados fornecidos
data = [[0, 0], [1, 1], [2, 2], [3, 4], [5, 6]]

# criar uma instância da classe Plotter
plotter = Plotter(data)

# chamar o método plot da instância
plotter.plot()
```
